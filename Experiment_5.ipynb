Using Python language do the exploratory data analysis of dataset imported in the lab 4.

Exploratory Data Analysis (EDA) involves analyzing and summarizing datasets using visual and statistical techniques to understand their structure, patterns, and insights. Below is an example of how to conduct EDA using Python with Pandas, NumPy, Matplotlib, and Seaborn libraries.

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
from google.colab import files
```

```python
upload=files.upload()
```

```python
import io
```

```python
df = pd.read_csv(io.BytesIO(upload['customers-100.csv']))
```

```python
df.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Columns', fontsize=16)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title('Boxplot of Numerical Columns', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Country', data=df, order=df['Country'].value_counts().index)
plt.title('Top Countries', fontsize=16)
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='City', data=df, order=df['City'].value_counts().index[:10])
plt.title('Top 10 Cities', fontsize=16)
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Company', data=df, order=df['Company'].value_counts().index[:10])
plt.title('Top 10 Companies', fontsize=16)
plt.show()
```
